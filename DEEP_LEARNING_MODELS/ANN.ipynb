```python
import tensorflow as tf
```

```python
# import some basic libraries
import numpy as np 
import pandas as pd
 
```

```python
# seprating feature and target label
dataset = pd.read_csv('Churn_Modelling.csv')
X = dataset.iloc[:,3:13]
y =dataset.iloc[:,13]

```

```python
X.head(),y.head()
```

```python
#Creating dummy Variables for converting categorical features to binary
geography = pd.get_dummies(X["Geography"],drop_first =True)
gender = pd.get_dummies(X["Gender"], drop_first=True)
```

```python
! pip install scikit-learn
```

```python
#Concatenate data frames
X= pd.concat([X,geography,gender],axis=1)
# drop unnecessary columns
X=X.drop(['Geography','Gender'],axis=1)

#split the datset into training and test set
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Feature Scaling - it is done  when we use gradient descent and distance problems to tackle noise and outliers

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
# till now we have prepared our dataset now lets make ANN
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LeakyReLU,PReLU,ELU 
from tensorflow .keras.layers import Dropout
```

```python
# initializing the ANN
classifier = Sequential()
```

```python
# adding the input and first hidden layer
classifier.add(Dense(units=11,activation='relu'))
```

```python
# adding the input and second hidden layer
classifier.add(Dense(units=6,activation='relu'))
```

```python
# adding the  output layer
classifier.add(Dense(units=1,activation='sigmoid'))
```

```python
classifier.compile(optimizer='adam',loss ='binary_crossentropy',metrics=['accuracy'])
```

```python
# early stopping - it is used when accuracy is increasing negligible.
import tensorflow as tf
early_stopping= tf.keras.callbacks.EarlyStopping(
    monitor = "val_loss",
    min_delta= 0.0001,
    patience =20,
    verbose=1,
    mode='auto',
    baseline = None,
    restore_best_weights=False
)
```

```python
model_history = classifier.fit(X_train,y_train,validation_split=0.33,batch_size=10,epochs=50,callbacks=early_stopping)
```

```python
# Making the prediction and evaluating the model

y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
# Calculate the Accuracy
from sklearn.metrics import accuracy_score
score=accuracy_score(y_pred,y_test)
```
