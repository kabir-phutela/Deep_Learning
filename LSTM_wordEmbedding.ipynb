```python
# Import necessary libraries

import numpy as np
import pandas as pd

# Keras tools for NLP and deep learning
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
# Simple dataset

texts = [
    "I love this movie",
    "This movie is good",
    "I hate this movie",
    "This movie is bad"
]

labels = [1,1,0,0]  # 1 = Positive, 0 = Negative
```

```python
# Create tokenizer object

tokenizer = Tokenizer()

# Learn vocabulary from texts

tokenizer.fit_on_texts(texts)

# Convert sentences into numbers

sequences = tokenizer.texts_to_sequences(texts)

print(sequences)
```

```python
# Make all sequences same length

X = pad_sequences(sequences, maxlen=5)

print(X)
```

```python
# Convert labels to array

y = np.array(labels)
```

```python
# Create model

model = Sequential()

# Embedding Layer
model.add(
Embedding(
input_dim=100,
output_dim=8,
input_length=5
)
)
```

```python
# Add LSTM layer

model.add(LSTM(8))
```

```python
# Output layer

model.add(Dense(1, activation='sigmoid'))
```

```python
# Prepare model for training

model.compile(
loss='binary_crossentropy',
optimizer='adam',
metrics=['accuracy']
)
```

```python
# Train model

model.fit(X,y,epochs=20)
```

```python
model.summary()
```

```python
# Test sentence

test = ["This movie is worst"]

# Convert to numbers

seq = tokenizer.texts_to_sequences(test)

# Padding

pad = pad_sequences(seq,maxlen=5)

# Prediction

pred = model.predict(pad)

print(pred)
```

```python
if pred > 0.5:
    print("Positive")
else:
    print("Negative")
```
